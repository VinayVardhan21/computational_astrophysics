```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("SCPUnion2_mu_vs_z.txt",
                    dtype=[("name", "S6"), ("z", "f8"), ("mu", "f8"), ("dz", "f8")])
```

```python
fig, ax = plt.subplots()
ax.errorbar(data["z"], data["mu"], yerr=data["dz"], fmt="o", ms=4)
```
